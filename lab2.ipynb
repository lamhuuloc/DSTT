```python
#baitap3.2
import numpy as np
A = np.reshape(np.arange(36.0), (6, 6))
print("Ma trận A:")
print(A)
I6 = np.identity(6)
print("\nMa trận đơn vị I6:")
print(I6)
print("\nSố phần tử trong A:", A.size)
print("Đường chéo chính của A:", np.diagonal(A))
A = A + I6
print("\nMa trận A sau khi cộng với I6:")
print(A)
vecB = np.array([1., 2., 3., 4., 5., 6.])
C = A.dot(vecB)
print("\nTích A * vecB (C):")
print(C)
D = np.array([[1., 2., 3., 4., 5., 6.],
              [1., 0., 1., 0., 1., 0.]])
print("\nMa trận D (2x6):")
print(D)
E = A.dot(D.T)  
print("\nMa trận E = A * D.T:")
print(E)
F = np.array([
    [1., 1.],
    [2., 0.],
    [3., 1.],
    [4., 0.],
    [5., 1.],
    [6., 0.]
])
print("\nMa trận F (6x2):")
print(F)
G = A.dot(F)
print("\nMa trận G = A * F:")
print(G)
A_inv = np.linalg.inv(A)
print("\nMa trận nghịch đảo của A:")
print(A_inv)
print("\nnp.linalg.inv(np.linalg.inv(A)) có bằng A không?")
print(np.allclose(np.linalg.inv(np.linalg.inv(A)), A))
```

```python
#bài tập 3
import numpy as np

def fib_matrix(k):
    F = np.array([[1, 1], [1, 0]], dtype=object)
    return np.linalg.matrix_power(F, k)

def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n+1):
        fib.append(fib[-1] + fib[-2])
    return fib

# Ví dụ với k = 7
k = 7
Fk = fib_matrix(k)
fib_seq = fibonacci(k + 1)

print("F_k matrix:")
print(Fk)

print(f"Fibonacci sequence up to F_{k}:")
print(fib_seq)

print(f"Fk contains: Fib_{k+1}={fib_seq[k+1]}, Fib_{k}={fib_seq[k]}, Fib_{k-1}={fib_seq[k-1]}")
print("So Fk = [[Fib(k+1), Fib(k)], [Fib(k), Fib(k-1)]]:", Fk[0,0]==fib_seq[k+1], Fk[0,1]==fib_seq[k])
```
